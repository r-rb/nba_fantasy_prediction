```python
# Discrete
disc = ['name','position','home_game','team_against','player_team']

# Continuous 
cont = ['ast','blk','min', 'pts','reb','stl','to','fantasy_points','min_points', \
            'max_points', 'ave_points_team', 'ave_points_vs_team']

train_test =['2014-2015','2015-2016','2016-2017']

hold_out = ['2017-2018']
with open('../data/final_data.pkl', 'rb') as fp:
    input_data = pickle.load(fp)
```

```python
# Remove undesirables
df = pd.DataFrame.from_dict(input_data)
df = df.drop(df[df.ave_points_team < 0.1].index)
df = df.drop(df[df.ave_points_vs_team < 0.1].index)
df.head()
```

```python
# Split into different shit
df_train = df[df.season.isin(train_test)]
df_train.head()

df_holdout = df[df.season.isin(hold_out)]
df_holdout.head()
```
